```python
import functions_model as f
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
k = 1. # emt rate
c = 2. # cooperativity

# cell adhesion coefficients
AE, A1, A2, A3, Am = 1., 0.75, 0.5, 0.25, 0.
pE1, pE2, pE3, p11, p12, p13, p22, p23, p33 = 1-AE*A1, 1-AE*A2, 1-AE*A3, 1-A1*A1, 1-A1*A2, 1-A1*A3, 1-A2*A2, 1-A2*A3, 1-A3*A3
```

```python
tm, e, h1, h2, h3, m = f.solve_model(c, k, pE1, pE2, pE3, p11, p22, p33, p12, p13, p23, two_dim=True)
```

Temporal trajectory

```python
plt.plot(tm, e, 'g-', label='E')
plt.plot(tm, h1, '-', color='yellow', label='H1')
plt.plot(tm, h2, '-', color='orange', label='H2')
plt.plot(tm, h3, '-', color='red', label='H3')
plt.plot(tm, m, '-', color='darkred', label='M')
plt.legend(loc='upper right')
plt.xlabel('Time (AU)')
plt.ylabel('Cell fraction')
```

Steady state cluster size distribution

```python
s = np.arange(1,101,1)
sum, clst_list = f.cluster_sum_ss(s.size, c, e[-1], h1[-1], h2[-1], h3[-1], m[-1], pE1, pE2, pE3, p11, p22, p33,
										p12, p13, p23, two_dim=True)
dist = f.find_dist(clst_list)
plt.plot( s, dist, 'o-' )
plt.xlim([1,10])
plt.xlabel('Cluster size (# cells)')
plt.ylabel('Probability')
```
